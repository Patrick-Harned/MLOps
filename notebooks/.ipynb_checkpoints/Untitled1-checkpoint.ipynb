```bash
cpdctl config users set qa-user --username=admin --password=password
```

```bash
cpdctl config users list
```

```bash
cpdctl config prif
```

```bash
cpdctl config profiles set qa-cluster --user qa-user --url https://zen-cpd-zen.apps.pwh.ocp.csplab.local
```

```bash
cpdctl config contexts set qa-context qa-cluster
```

```bash
cpdctl spaces list | grep prod-space | awk 'END{print NR}'
```

```bash
cpdctl spaces list 
```

```bash
str1=$(cpdctl spaces list | grep prod-space | awk 'END{print NR}')
if [[ $str1 != 1 ]]; then
 echo "deleting extra spaces";
 cpdctl spaces list | grep prod-space | awk 'NR>1{print $1}' | xargs -I{} cpdctl spaces delete --space_id {}
fi
if [[ $str1 == 0 ]]; then
 echo "creating space"
cpdctl spaces create --name prod-space
 else
   cpdctl spaces list
   export SPACE_ID=$(cpdctl spaces list | grep prod-space | awk '{ print $1}')
   echo $SPACE_ID
fi
```

```bash
models=$(cpdctl wml models list --space-id $SPACE_ID | awk 'NR>1')
if [[ $models == "Nothing to show." ]]; then
   echo "No models found, creating model"
   cpdctl wml models create --name mlopsmodel --space-id $SPACE_ID --software_spec '{"name": "scikit-learn_0.22-py3.6"}' --type 'scikit-learn_0.22'
   export MODEL_ID=$(cpdctl wml models list --space-id $SPACE_ID | grep mlopsmodel | awk '{ print $1}')
 else
   cpdctl wml models list --space-id $SPACE_ID
   export MODEL_ID=$(cpdctl wml models list --space-id $SPACE_ID | grep mlopsmodel | awk '{ print $1}')
   echo $MODEL_ID
   cpdctl wml models upload-content --body mlopsmodelpipeline.tar.gz --model_id $MODEL_ID --content_format 'binary' --space-id $SPACE_ID --output json
   
fi
```

```bash
deployments=$(cpdctl wml deployments list --space-id $SPACE_ID |awk 'NR>1')
echo $deployments
if [[ $deployments == "Nothing to show." ]]; then
echo "Creating Deployment"
cpdctl wml deployments create --name mlopstestdeploy --online '{"description": "mlopsdeploy"}' --asset '{"id": "'"$MODEL_ID"'"}' --space-id $SPACE_ID
fi
DEPLOY_ID=$(cpdctl wml deployments list --space-id $SPACE_ID | awk 'NR>2{print $1}')
echo $DEPLOY_ID
```
